```python
import numpy as np
data = np.genfromtxt('aug_train.csv', delimiter = ',', dtype = str , encoding = 'utf-8')

print('Full shape:',data.shape)
print('Total data points : ', data.shape[0] * data.shape[1])
print('\nColumn headers: ')
print(data[0])
print('\nFirst 3 rows of actual data:')
print(data[1:4])
print('\nLast 3 rows:')
print(data[-3:])
```

```python
header = data[0]
raw= data[1:]

print('Headers:',header)
print('\nData shape (no header):', raw.shape)
print('\nTotal candidates:',raw.shape[0])
print('Total features:',raw.shape[1])

experience_col = raw[:,8]
print('\nExperience column (first 20 values)')
print(experience_col[:20])

unique_exp = np.unique(experience_col)
print('\nAll unique experience values:')
print(unique_exp)
print('Total unique values:', len(unique_exp))
```

```python
print("Missing values per column:")
print("-"*40)

for i,col_name in enumerate(header):
    col=raw[:,i]
    missing = np.sum(col=='')
    missing_pct = (missing / raw.shape[0])*100
    print(f"{col_name:<30} missing: {missing:>5} ({missing_pct:.1f}%)")
```

```python
cleaned = raw.copy()

for col_idx in [3,9,10]:
    col=cleaned[:,col_idx]
    col[col=='']='Unknown'
    cleaned[:,col_idx]=col

cleaned[:,7][cleaned[:,7]=='']= 'No Major'

experience_col = cleaned[:,8]
mask = experience_col !=''
cleaned = cleaned[mask]

print("Original rows:", raw.shape[0])
print('Cleaned rows:', cleaned.shape[0])

print('\nAfter cleaning:')
for idx,name in [(3,'gender'),(7,'major_discipline'),(8,'experience'),(9,'company_size'),(10,'company_type')]:
    missing = np.sum(cleaned[:,idx]=='')
    print(f"{name:<25} missing:{missing}")
```

```python
exp = cleaned[:,8].copy()

print('Before cleaning:')
print('Unique values:', np.unique(exp))
print('dtype:',exp.dtype)

exp[exp=='<1']='0'
exp[exp=='>20']='21'
exp_numeric = exp.astype('float')
print('\nAfter cleaning')
print('dtype:', exp_numeric.dtype)
print('Min experience:',np.min(exp_numeric))
print('Max Experience:',np.max(exp_numeric))
print('Mean Experience:',np.mean(exp_numeric).round(2))
print('Median Expereince:',np.median(exp_numeric))

brackets = [(0,2),(2,5),(5,10),(10,15),(15,22)]
print('\nExperience distribution:')
for low,high in brackets:
    count=np.sum((exp_numeric>=low) &(exp_numeric<high))
    pct = (count/len(exp_numeric))*100
    bar  = '█' *int(pct)
    print(f'{low:>2}    -{high:<3}yrs: {count:>5} candidates {bar} {pct:.1f}%')
```
